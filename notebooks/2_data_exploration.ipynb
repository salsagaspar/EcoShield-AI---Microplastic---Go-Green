# Phase 2: Exploratory Data Analysis (EDA)

This notebook aims to explore the datasets cleaned and merged during Phase 1. The exploration is divided into three parts corresponding to the three pillars of the **EcoShield AI** platform:
1. **Pillar 1: Microplastic Distribution & Correlation Analysis**
2. **Pillar 2: Waste Cleanup Efficiency & Success Factors**
3. **Pillar 3: Green Initiatives Financial & Policy Impact Analysis**

We will use descriptive statistics and data visualizations to discover key insights for downstream ML modeling.

## 1. Setup Environment & Load Cleaned Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["font.size"] = 12

# Cleaned data directory
data_dir = "../data/"

df_micro = pd.read_csv(os.path.join(data_dir, "cleaned_microplastics.csv"))
df_waste = pd.read_csv(os.path.join(data_dir, "cleaned_waste_events.csv"))
df_green = pd.read_csv(os.path.join(data_dir, "cleaned_green_initiatives.csv"))

print("Three joined datasets loaded successfully.")
```

## 2. Descriptive Statistical Analysis

```python
print("=== MICROPLASTIC SURVEY METRICS ===")
display(df_micro[['concentration_particles_per_l', 'size_um', 'ph_level', 'turbidity_ntu', 'salinity_ppt', 'pollution_index', 'biodiversity_score']].describe())

print("\n=== WASTE CLEANUP OPERATION METRICS ===")
display(df_waste[['weight_collected_ton', 'volume_collected_m3', 'microplastic_fraction_pct', 'cost_usd', 'co2_saved_kg', 'num_volunteers', 'recycling_rate_pct']].describe())

print("\n=== GREEN INITIATIVE PROJECT METRICS ===")
display(df_green[['total_budget_usd', 'plastic_removed_ton', 'co2_equivalent_saved_ton', 'beneficiaries_reached', 'media_coverage_count', 'policy_influence_score']].describe())
```

## 3. Pillar 1: Microplastic Distribution & Relationships

```python
# Plot 1: Target Variable Distribution - Microplastic Concentration
plt.figure(figsize=(10, 5))
sns.histplot(df_micro["concentration_particles_per_l"], kde=True, color="teal", bins=50)
plt.title("Distribution of Microplastic Concentration (Particles / Liter)")
plt.xlabel("Concentration")
plt.ylabel("Frequency")
plt.show()

# Plot 2: Concentration by Plastic Type and Location Type
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
sns.boxplot(data=df_micro, x="plastic_type", y="concentration_particles_per_l", ax=axes[0], palette="viridis")
axes[0].set_title("Microplastic Concentration by Plastic Type")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)

sns.boxplot(data=df_micro, x="location_type", y="concentration_particles_per_l", ax=axes[1], palette="magma")
axes[1].set_title("Microplastic Concentration by Location Type")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Pearson Correlation Heatmap for Environmental Variables
plt.figure(figsize=(10, 8))
cols = ["concentration_particles_per_l", "size_um", "ph_level", "turbidity_ntu", "salinity_ppt", "depth_m", 
        "annual_rainfall_mm", "avg_temperature_c", "pollution_index", "biodiversity_score"]
corr = df_micro[cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Environmental Features & Microplastics")
plt.show()
```

## 4. Pillar 2: Cleanup Operations & Efficiency Factors

```python
# Plot 1: Cost vs Weight Collected Scatter Plot by Weather Condition
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_waste.sample(1000, random_state=42), x="cost_usd", y="weight_collected_ton", 
                hue="weather_condition", alpha=0.7, palette="Set2")
plt.title("Relationship of Budget Cost (USD) vs Waste Collected (Tons)")
plt.xlabel("Cost (USD)")
plt.ylabel("Weight Collected (Tons)")
plt.xscale("log")
plt.yscale("log")
plt.show()

# Plot 2: Average Waste Collected by Waste Source
plt.figure(figsize=(12, 5))
order = df_waste.groupby("waste_source")["weight_collected_ton"].median().sort_values(ascending=False).index
sns.barplot(data=df_waste, x="waste_source", y="weight_collected_ton", order=order, palette="Blues_r")
plt.title("Average Waste Collected by Industry Waste Source")
plt.xticks(rotation=45)
plt.show()

# Plot 3: Number of Volunteers vs CO2 Saved
plt.figure(figsize=(10, 5))
sns.regplot(data=df_waste.sample(500, random_state=42), x="num_volunteers", y="co2_saved_kg", 
            scatter_kws={'alpha':0.5, 'color':'forestgreen'}, line_kws={'color':'red'})
plt.title("Impact of Volunteer Counts on CO2 Saved (kg)")
plt.show()
```

## 5. Pillar 3: Green Initiatives & Policy Outcomes

```python
# Plot 1: Total Budget vs Plastic Removed Scatter Plot by Project Impact Level
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_green.sample(1000, random_state=42), x="total_budget_usd", y="plastic_removed_ton", 
                hue="impact_level", alpha=0.7, palette="viridis")
plt.title("Project Budget Impact on Total Plastic Waste Removed (Tons)")
plt.xlabel("Budget (USD)")
plt.ylabel("Plastic Removed (Tons)")
plt.xscale("log")
plt.yscale("log")
plt.show()

# Plot 2: Policy Outcomes Stacked Bar Chart by Initiative Type
plt.figure(figsize=(12, 6))
df_policy_cross = pd.crosstab(df_green["initiative_type"], df_green["policy_outcome"], normalize="index") * 100
df_policy_cross.plot(kind="bar", stacked=True, cmap="Set3", ax=plt.gca())
plt.title("Stacked Distribution of Policy Outcomes by Initiative Type")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plot 3: Policy Influence Score by SDG Target Focus
plt.figure(figsize=(12, 5))
sns.boxplot(data=df_green, x="primary_sdg_focus", y="policy_influence_score", palette="coolwarm")
plt.title("Organization Policy Influence Score by Primary SDG Target Focus")
plt.show()
```

## 6. Key Conclusions for AI Modeling

From the exploratory data analysis, we gather the following key inputs for model building:
1. **Microplastic Predictive Model:** Spatial features (coordinates, location types) and physical-chemical values (turbidity, salinity) hold strong correlations and boxplot variance with concentration. This confirms regression trees (XGBoost/LightGBM) will perform exceptionally well.
2. **Cleanup Operations Optimizer:** Cost and volunteer count are primary operational inputs. Weather conditions significantly impact cleanup weight (with stormy/rainy weather lowering yields), which validates the need for weather-aware resource routing.
3. **Policy Lobbying Success Classifier:** High policy outcomes ("Major Bill Passed") are strongly skewed towards *Policy Lobbying* and *Tech Innovation* initiative types, but are also dependent on the organizing body's overall `policy_influence_score` rather than budget alone.
